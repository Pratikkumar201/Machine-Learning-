```python
# Percentage of companies got investment or not ?
got_in=data[data['deal_amount']>0.0].shape
not_in=data[data['deal_amount']==0.0].shape
print("The number of companies who got the Investment: ",got_in[0])
print("The number of companies who didn't got the Investment: ",not_in[0])
print("\n---------\n")
list1=["Got Investment", "Not received Investment"]
plt.pie([got_in[0],not_in[0]],labels=list1,autopct='%.1f%%')
plt.title('Number of companies which got Investment or not: ')
plt.show()
```

```python
data[data['total_sharks_invested']==5]
```

```python
total_shark=data[data['total_sharks_invested']==5]
figure=px.bar(total_shark,x='brand_name',y='deal_amount',title="Five shark Deal Brands and the total investment ",template='plotly_dark')
figure.show()
```

```python
#  Highest pitch ask amount
high=data[data['pitcher_ask_amount']>100]
figure=px.bar(high,x='brand_name',y='pitcher_ask_amount',title='Highest Pitch Ask Amount:',text_auto=True,template='plotly_dark')
figure.show()
```

```python
# Ask Equity and Deal Equity of Highest Pitch Ask Amount Brand
figure=px.bar(high,x='brand_name',y='ask_equity',title="Ask Equity and Deal Equity of Highest Pitch Ask Amount Brand",template='plotly_dark')
figure.show()
```
